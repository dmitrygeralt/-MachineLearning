## Обрабатываем пропущенные значения

Если данные содержат пропущенные значения, то имеется две простые альтернативы:
- удалить столбцы с такими значениями (`data = data.dropna(axis = 1)`),
- удалить строки с такими значениями (`data = data.dropna(axis = 0)`).

После этого данных может стать совсем мало.

Куда более эффективным (и в то же время по-прежнему простым) методом обычно является заполнение пропущенных значений для каждой количественной переменной их средним или медианным значением. 
Для категориальных признаков пропущенные значения можно заполнить самым популярным значением в столбце.

```python
data.count(axis=0)
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
data.count(axis=0)
```

Можно увидеть, что строк с пропущенными категориальными признаками не имеется.

```python
data.loc[(data['Income'] <= 10000)] = 10000
data.loc[(data['Income'] > 10000) & (data['Income'] < 20000)] = 20000
data.loc[(data['Income'] > 20000) & (data['Income'] < 30000)] = 30000
data.loc[(data['Income'] > 30000) & (data['Income'] < 40000)] = 40000
data.loc[(data['Income'] > 40000) & (data['Income'] < 50000)] = 50000
data.loc[(data['Income'] > 50000) & (data['Income'] < 60000)] = 60000
data.loc[(data['Income'] > 60000) & (data['Income'] < 70000)] = 70000
data.loc[(data['Income'] > 70000) & (data['Income'] < 80000)] = 80000
data.loc[(data['Income'] > 80000) & (data['Income'] < 90000)] = 90000
data.loc[(data['Income'] > 90000) & (data['Income'] < 100000)] = 100000
data.loc[(data['Income'] >= 100000)] = 150000
```

## Векторизация

Нужно что-то сделать с категориальными (номинальными) признаками

К небинарными признакам применим метод _векторизации_, 
который заключается в следующем.

Признак `j`, принимающий `s` значений, заменим на `s` признаков, принимащих значения `0` или `1`,
в зависимости от того, чему равно значение исходного признака `j`.


Применим метод векторизации (Векторизация категориальных признаков)

Будет сделано следующее:

Для небинарных категориальных признаков:
Вместо N значений признака будет использоваться N признаков со значением 0 или 1 (в зависимости от значения исходного признака), таким образом, получив вектор сумма элементов которого всегда равна 1 (в силу того, что пропуски уже устранены)
Для бинарных категориальных признаков:
Вместо категории будем использовать число - 0 или 1 соответственно

# Подбор гиперпараметров

 Теперь попробуем классификатор с большим числом параметров

### Подберем оптимальное значение параметра регуляризации alpha

```python
alpha_arr = np.logspace(-3, 2, 50)
test_err = []
train_err = []
for alpha in alpha_arr:
    perc = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    perc.fit(x_train, y_train)

    y_train_pred = perc.predict(x_train)
    y_test_pred = perc.predict(x_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'test')
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'train ')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

Видно,для alpha :
* [0.001;0.004] - ошибка на тестовой выборке меньше ошибки на обучающей выборке,но растёт.
* [0.004;1] - больше ошибки на обучающей выборки  => имеем переобучение.
* При alpha > 1 эффект переобучния пропадает (ошибка на тестовой выборке снова меньше ошибки на обучающей)

Выведем минимальное значение ошибки:

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

Выведем оптимальное значение alpha:

```python
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

```python
alpha_opt = alpha_opt[0]
print("Оптимальное значение параметра регуляризации alpha = ",alpha_opt)   
```

###  Классификатор при найденном оптимальном значении alpha:

Повторим обучение:

```python
perc = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (100,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
perc.fit(x_train, y_train)

y_train_pred = perc.predict(x_train)
y_test_pred = perc.predict(x_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

```python
from sklearn import decomposition 
pca = decomposition.PCA(n_components = 10)
X_train_reduc = pca.fit_transform(X_train)
X_test_reduc = pca.transform(X_test)
```

 Обучим классификатор вида Gradient Boosted Trees. Число деревьев = 100

```python
from sklearn import ensemble
gbt_model = ensemble.GradientBoostingClassifier(n_estimators = 100, max_depth = 2, learning_rate= 0.4)
gbt_model.fit(X_train_reduc, y_train)
```

 Посчитаем и выведем ошибку на обучающей выборке

```python
err_train = np.mean(y_train != gbt_model.predict(X_train_reduc))
print(err_train)
```

Посчитаем и выведем ошибку на тестовой выборке

```python
err_test  = np.mean(y_test  != gbt_model.predict(X_test_reduc))
print(err_test)
```

Наблюдаем переобучение, т.к ошибка на обучающей выборке составляет 0,5%, а на тестовой 11,5% !

 Построим график зависимости ошибок (ошибка на обучающей выборке + ошибка на тестовой выборке) от количества компонент в методе

```python
components_arr = np.arange(1, 30, 1)
test_err = []
train_err = []
oob_err = []
for components in components_arr:
    pca = decomposition.PCA(n_components = components)
    X_pca_train = pca.fit_transform(X_train)
    X_pca_test = pca.transform(X_test)
    gbt_model = ensemble.GradientBoostingClassifier(n_estimators = 100, max_depth = 2, learning_rate= 0.4)
    gbt_model.fit(X_pca_train, y_train)
    train_err.append(np.mean(y_train != gbt_model.predict(X_pca_train)))
    test_err.append(np.mean(y_test  != gbt_model.predict(X_pca_test)))
```

```python
plt.plot(components_arr, train_err, 'b-o', label = 'train')
plt.plot(components_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(components_arr), np.min(components_arr)])
plt.title('Error vs. Num of components')
plt.xlabel('Num of components')
plt.ylabel('Error')
plt.legend()
pass
```

Из графика видно,что наименьшая ошибка достигается при числе компонет от 3-х до 5-ти. При числе же компонент,больше 5 ошибка не убывает (только местами растёт). Следовательно,лучше взять число компонент от 3-х до 5. Посмотрим,что скажут вычисления.
